```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('stepCount.csv')
df
```

```python
# convert the create_time to datetime
df['create_time'] = pd.to_datetime(df['create_time'])
df
```

```python
# conver create_time to month
df['month'] = df['create_time'].dt.month
df
```

```python
# remove other months other than month 7
df_7 = df[df['month'] == 7]
df_7.head(4)
```

```python
# convert create_time to weekdays with name
df_7['weekday'] = df_7['create_time'].dt.day_name()
df_7
```

```python
# first day of the week is Monday
df_7['weekday'] = df_7['weekday'].map({'Monday': 1, 'Tuesday': 2, 'Wednesday': 3, 'Thursday': 4, 'Friday': 5, 'Saturday': 6, 'Sunday': 7})
df_7
```

```python
# plot the step count per day
df_7.groupby('weekday').sum()['count'].plot(kind='bar')

#rename 1 to Monday, 2 to Tuesday, 3 to Wednesday, 4 to Thursday, 5 to Friday, 6 to Saturday, 7 to Sunday
df_7['weekday'] = df_7['weekday'].map({1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'})

plt.show()
```

```python
# heatmap
sns.heatmap(df_7.groupby('weekday').sum()['count'].reset_index().pivot('weekday',  'count'), annot=True, cmap='RdBu')
plt.show()
```
